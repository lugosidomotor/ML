```
df = pd.read_table('verified_pb.data', delim_whitespace=True, names=["Sex", "SPKR", "Phoneme-Number", "Phoneme-Ascii","F0", "F1", "F2", "F3" ])
df.head(10)
```

```
#CLEAN DATA FROM *
df['Phoneme-Ascii'] = df['Phoneme-Ascii'].str.replace('*', '')
df.head(10)
```

```
#GET THE FIRST 4 MALE, FIRST 4 FEMALE AND FIRST 3 CHILD

male = []
female = []
child = []

for row in df.iterrows():
  if row[1][0] == 1 and row[1][1] not in male and len(male) < 4:
    male.append(row[1][1])

for row in df.iterrows():
  if row[1][0] == 2 and row[1][1] not in female and len(female) < 4:
    female.append(row[1][1])
  
for row in df.iterrows():
  if row[1][0] == 3 and row[1][1] not in child and len(child) < 2:
    child.append(row[1][1])


test = male + female + child
print(test)
```

```
#CREATE TEST SET
df_test =  df.loc[df['SPKR'].isin(test)]
```

```
#CREATE TRAIN SET
df_train = pd.concat([df, df_test]).drop_duplicates(keep=False)
```

```
#TARGET
X_train = df_train.iloc[:,2].values
X_test = df_test.iloc[:,2].values

#FEATURE
Y_train = df_train.iloc[:, 4:8].values
Y_test = df_test.iloc[:, 4:8].values

#TRAIN THE MODEL
tree = DecisionTreeClassifier().fit(X_train.reshape(-1, 1), Y_train)

#PREDICTION
tree_prediction = tree.predict(X_test.reshape(-1, 1))

#PREDICTED VALUES
print(tree_prediction)
#REAL VALUES
Y_test
```
